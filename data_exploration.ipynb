```python
from pysam import FastaFile
 
fasta = "orfs_out/orfs.fa"
# read FASTA file
sequences_object = FastaFile(fasta)
```

```python
chrom_segments = {}

for i in range(len(sequences_object.references)):
    chrom_segments[sequences_object.references[i]] = [(0, sequences_object.lengths[i])]
    sequences_object.close()
```
